```python
%reset -f
from ROOT import *
from WaveReco import WaveReco as WR
%matplotlib widget
%jsroot on
```

```python
runName = "run650341"
entries = -500
```

```python
wr = WR.WaveReco()

wr.reset()
wr.setPaths(runName, "data/roottople/{}.root", "data/recod/{}.root", "data/recod_ana/{}.root", "t", "t")
wr.config(setMaxEvts=entries, setSpecsToSave=1000, setSaveEventNumber="eventNumber")

wr.addDigi(wr.Digi())
digi = wr.digis[-1]

wr.wboxes = []
```

```python
cindyin, cindyout =     [F"wave{ii}" for ii in [6,7]],                          "cindy{}"
fastin, fastout =       [F"wavefast{ii}" for ii in [0,8,16,24,32,40,48,56]],    "sync{}"
ciaxin, ciaxout =       [F"wave{ii}" for ii in [9, 10]],                        ["ciaxf", "ciaxb"]
criin, crout =          [F"wave{ii}" for ii in [18, 19, 10+32, 11+32]], [F"center_{ii}" for ii in ["sl", "sr", "pl", "pr"]]

cindy =     wr.Wbox(wr.digis[-1], chin=cindyin, chout=cindyout, saveW=True, inv=True, 
                    cf=0.06, thr=45.0, pkMsk=[0.05,0.5], bRng=[-15,-4.5], qRng=[-4,15], spRng=[-8,10], pkCut=[None, None])

ciax =      wr.Wbox(wr.digis[-1], chin=ciaxin, chout=ciaxout, saveW=True, inv=True, 
                    cf=0.08, thr=45.0, pkMsk=[0.05,0.5], bRng=[-19,-9], qRng=[-8.5,12.5], spRng=[-12,6], pkCut=[None, None])

crilin =    wr.Wbox(wr.digis[-1], chin=criin, chout=crout, saveW=True, inv=False, 
                    cf=0.12, thr=45.0, pkMsk=[0.05,0.5], bRng=[-24,-12], qRng=[-11.5, 38], spRng=[-11.5,6], pkCut=[6.0, None])

sync =      wr.Wbox(wr.digis[-1], chin=fastin, chout=fastout, saveW=True, inv=True, 
                    cf=0.12, thr=45.0, pkMsk=[0.05,0.5], bRng=[-24,-12], qRng=[-11.5, 38], spRng=[-11.5,6], pkCut=[6.0, None])

if (runName == ""):
    pass

elif (runName == "run650333" or runName == "run650332" or runName == "run650316" or runName == "run650309"):
    digi.timBin = 0.2
    digi.compute()
    
    # wr.addWBox(cindy)
    wr.addWBox(sync)
    wr.addWBox(crilin)
    
elif (runName == "run650343" or runName == "run650342"  or runName == "run650341"):
    digi.timBin = 0.2
    digi.compute() 
    
    cindy.cfScan =  [0.03, 0.06, 0.09, 0.12, 0.15, 0.18]
    cindy.chin = [F"wave{ii}" for ii in [11,12]]
    cindy.compute()
    wr.addWBox(cindy)
    wr.addWBox(ciax)
```

```python
wr.launch()
```
